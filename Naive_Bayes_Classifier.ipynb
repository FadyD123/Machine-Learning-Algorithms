```python
# Naive Bayes Classifier

import pandas as pd # We will use pandas library for data preprocessing 
import matplotlib.pyplot as plt # we will visualise the data 
import numpy as np # We use the numpy library for linear algebrea
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris # We will load the dataset load_iris 
iris = load_iris() 
Data_iris = iris.data
Data_iris = pd.DataFrame(Data_iris, columns = iris.feature_names)
#iris.target
# The target part is the second part of the dataset
Data_iris['label'] = iris.target
# We combine the 2 parts into one dataframe called data_iris
#Data_iris
# Here we visualise the data
# We draw the petal length and petal width 
# we have 3 values for the target: 0 , 1 and 2
# Here we color the data points with respect to its target
plt.scatter(Data_iris.iloc[:,2], Data_iris.iloc[:,3], c=iris.target)
plt.xlabel('Petal Length cm')
plt.ylabel('Petal width cm')
plt.show()
x = Data_iris.iloc[:,0:4] # x is just the data
y = Data_iris.iloc[:,4] # y is the target
x
#y
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, train_size = 0.8, 
                                                    random_state = 88, shuffle = True, stratify = y)

NB = GaussianNB()

NB.fit(X_train, y_train)

y_pred = NB.predict(X_test)

from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)

from sklearn.model_selection import cross_val_score

Scores_NB = cross_val_score(NB, x, y, cv = 10)

Scores_NB
```
